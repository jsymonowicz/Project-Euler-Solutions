# Distinct Powers

## Problem 29

Consider all integer combinations of $a^{b}$ for $2 \le a \le 5$ and $2 \le b \le 5$:

<center>$2^{2} = 4$,&nbsp;&nbsp;&nbsp;&nbsp;$2^{3} = 8$,&nbsp;&nbsp;&nbsp;&nbsp;$2^{4} = 16$,&nbsp;&nbsp;&nbsp;&nbsp;$2^{5} = 32$</center>
<center>$3^{2} = 9$,&nbsp;&nbsp;&nbsp;&nbsp;$3^{3} = 27$,&nbsp;&nbsp;&nbsp;&nbsp;$3^{4} = 81$,&nbsp;&nbsp;&nbsp;&nbsp;$3^{5} = 243$</center>
<center> (...) &nbsp;&nbsp;&nbsp;&nbsp;$5^{5} = 3125$</center>

If they are then placed in numerical order, with any repeats removed, we get the following sequence of 15 distinct terms:<center>4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 256, 625, 1024, 3125</center>


How many distinct terms are in the sequence generated by $a^{b}$ for $2 \le a \le 100$ and $2 \le b \le 100$?

```python
# Declare the set to store terms
distinct_terms = set()

# Define max a and b
max_a = 100
max_b = 100

# Calculate and store all a^b
for a in range (2, max_a + 1):
    for b in range (2, max_b + 1):
        term = a**b
        distinct_terms.add(term)

# Calculate number of terms in the set of a^b
distinct_terms_length = len(distinct_terms)

# Print out the result
print(distinct_terms_length)
```
